```python
# read text files in pieces
# when processing very large files or figuring out the right set of arguments to correctly process a large file
# read in a small piece of a file or smaller chunks of the file
pd.options.display.max_rows = 10
pd.read_csv('examples/ex4.csv')
# if you want to read a small number of rows(avoiding the entire file, specify that with nrows)
pd.read_csv('examples/ex4.csv', nrows=5)
```

```python
# to read a file in pieces, specify a chunksize as a number of rows
chunker = pd.read_csv('examples/ex4.csv', chunksize=1000)
chunker
# the TextParser object returned by read_csv allows you to iterate over the parts of the file according to the chunksize
tot = pd.Series([])
for piece in chunker:
    tot = tot.add(piece['key'].value_counts(), fill_value=0)
tot = tot.sort_values(ascending=False)
tot[:10]
```
