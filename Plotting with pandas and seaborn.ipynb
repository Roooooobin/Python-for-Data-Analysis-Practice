```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# seaborn, a statistical graphics library. Seaborn simplifies creating many common visualization types
```

```python
# line plots
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
print(df)
df.plot()
# df.plot() is equivalent to df.plot.line()
```

```python
# bar plots
# the plot.bar() and plot.barh() make vertical and horizontal bar plots
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

```python
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
print(df)
df.plot.bar()
# Note that the name “Genus” on the DataFrame’s columns is used to title the legend
# We create stacked bar plots from a DataFrame by passing stacked=True, resulting in the value in each row being stacked together
df.plot.barh(stacked=True, alpha=0.5)
```

```python
# suppose we wanted to make a stacked bar plot showing the percentage of data points for each party size on each day
tips = pd.read_csv('examples/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
print(party_counts)
party_counts = party_counts.loc[:, 2:5]  # filter 1- and 6-person parties
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
print(party_counts.sum(1))
print(party_pcts)
party_pcts.plot.bar()
```
