```python
# categorical data
# background and motivation
"""
frequently, a column in a table may contain repeated instances of a smaller set of distinct values
unqiue and value_counts
"""
values = pd.Series(['apple', 'orange', 'apple', 'apple'] * 2)
values
pd.unique(values)
pd.value_counts(values)
# a best practice is to use so-called dimension tables containing the distinct values and store the primary observations as integer keys
values = pd.Series([0, 1, 0, 0] * 2)
dim = pd.Series(['apple', 'orange'])
values
dim
# use the take method to restore the original Series of strings
dim.take(values)
# This representation as integers is called the categorical or dictionary-encoded representation
# The array of distinct values can be called the categories, dictionary, or levels of the data.
```

```python
# computations with categoricals
# pandas.qcut returns pandas.Categorical
np.random.seed(12345)
draws = np.random.randn(1000)
draws[:5]
bins = pd.qcut(draws, 4)
bins
bins = pd.qcut(draws, 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
bins
bins.codes[:10]
bins = pd.Series(bins, name='quartile')
results = (pd.Series(draws)
           .groupby(bins)
           .agg(['count', 'min', 'max'])
           .reset_index())
results
results['quartile']
"""
If you do a lot of analytics on a particular dataset, converting to categorical can yield
substantial overall performance gains. A categorical version of a DataFrame column
will often use significantly less memory, too.
"""
```

```python
# categorical methods
s = pd.Series(['a', 'b', 'c', 'd'] * 2)
cat_s = s.astype('category')
cat_s
cat_s.cat.codes  # cat provides access to categorical methods
cat_s.cat.categories

actual_categories = ['a', 'b', 'c', 'd', 'e']
cat_s2 = cat_s.cat.set_categories(actual_categories)
cat_s2
cat_s3 = cat_s[cat_s.isin(['a', 'b'])]
cat_s3.cat.remove_unused_categories()
```

```python
# advanced groupby use
# group transforms and "unwrapped" groupbys
df = pd.DataFrame({'key': ['a', 'b', 'c'] * 4,
                   'value': np.arange(12.)})
df
g = df.groupby('key').value
g.mean()
"""
Suppose instead we wanted to produce a Series of the same shape as df['value'] but
with values replaced by the average grouped by 'key'. We can pass the function
lambda x: x.mean() to transform:
"""
g.transform(lambda x: x.mean())
g.transform('sum')
g.transform(lambda x: x * 2)
df['value'].apply(lambda x: x * 2)
df['value'] = df['value'].apply(lambda x: x * 2)
df
```

```python
# more complicated
g.transform(lambda x: x.rank(ascending=False))
def normalize(x):
    return (x - x.mean()) / x.std()
g.transform(normalize)
g.apply(normalize)
```

```python
# grouped time resampling
N = 15
times = pd.date_range('2017-05-20 00:00', freq='1min', periods=N)
df = pd.DataFrame({'time': times,
                   'value': np.arange(N)})
df
df.set_index('time').resample('5min').count()
# if an additional group key column
df2 = pd.DataFrame({'time': times.repeat(3),
                    'key': np.tile(['a', 'b', 'c'], N),
                    'value': np.arange(N * 3.)})
df2[:7]
# to do the same resampling for each value of 'key', we introduce the pandas.TimeGrouper
time_key = pd.Grouper(freq='5min')
resampled = (df2.set_index('time').groupby(['key', time_key]).sum())
```
